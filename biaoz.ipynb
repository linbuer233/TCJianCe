```python
import xarray as xr
import numpy as np
```

```python
ds=xr.open_dataset('../data/all.nc')
ds
```

```python
sp=ds['sp'].loc[ds.time.dt.day.isin(np.arange(1,31,1))].loc['2017-07-01':'2019-09-01',:,:]
sst=ds['sst'].loc[ds.time.dt.day.isin(np.arange(1,31,1))].loc['2017-07-01':'2019-09-01',:,:]
t=ds['t'].loc[ds.time.dt.day.isin(np.arange(1,31,1))].loc['2017-07-01':'2019-09-01',:,:,:]
u=ds['u'].loc[ds.time.dt.day.isin(np.arange(1,31,1))].loc['2017-07-01':'2019-09-01',:,:,:]
v=ds['v'].loc[ds.time.dt.day.isin(np.arange(1,31,1))].loc['2017-07-01':'2019-09-01',:,:,:]
vo=ds['vo'].loc[ds.time.dt.day.isin(np.arange(1,31,1))].loc['2017-07-01':'2019-09-01',:,:]
```

```python
vo
```

```python
xr.Dataset({'sp':sp,'sst':sst,'t':t,'u':u,'v':v,'vo':vo}).to_netcdf('../data/x_train.nc')
```
